```python
tracks = biggiesmalls_response['tracks']
```

```python
#counting problem

#initial condition (start value)

track_count = 0
explicit_track_count = 0

for track in tracks:
    print(track['name'], track['explicit'])
    track_count = track_count + 1;
    # if not track['explicit'] would retun the inverse!
    if track['explicit']:
        explicit_track_count = explicit_track_count + 1;
        print("The current track count:",explicit_track_count)
print("Overall, I discovered", track_count, "tracks")
print("Overall, I discovered", explicit_track_count, "explicit tracks")
if track_count > 0:
    print("Which means", 100 * explicit_track_count / track_count, "are explicit")
else:
    print("No top tracks found.")
```
