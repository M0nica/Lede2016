# Guide to metacharacters
* '.' is a wildcard for any one single character (even a blank space)
* \w any alphanumeric (a-z, A-Z, 0-9, _)
* \s any whitespace char (" ", \t, \n)
* \S any non-whitespace
* \d any digit (0-9)
* note: \. is an actual period

```python
# regex to find times in e-mail subject lines
[line for line in subjects if re.search("\d:\d\d\wm", line)]
#a digit followed by a colon , follwowed by another digit, followed by another digit, followd by a letter, followed by m
```

```python
[line for line in subjects if re.search("\.\.\.\.", line)]
```

```python
#subject liens that have dates, e.g.. 12/01/99
[line for line in subjects if re.search("\d\d/\d\d/\d\d", line)]
```

## meta characters 3: quantifiers

{n} matches exactly n times
{n,n} matches atleast ntimes, but no more than n times
{n, } matches atleast ntimes but maybe infinite times
+     matches at least once ({1,})
*     matches zero or more times (either find it or not)
?     matches one time or zero times

```python
[line for line in subjects if re.search(r"[A-Z]{15,}",line)] # atleast 15 uppercase letters in a row repeared
```

```python
[line for line in subjects if re.search(r"^F[wW]d?:",line)] # return the lines that may or may not have a d in fwd
```

```python
[line for line in subjects if re.search(r"[nN]ews.*!$",line)] # returns any line that has news anywhere
#followed by * any characters and then ends with an exclamation point 
```

```python
[line for line in subjects if re.search(r"^R[eE]:.*\b",line)]
```

```python
courses = [
    "CSCI 105: Introductory Programming for Cat-Lovers",
    "LING 214: Pronouncing Things Backwards",
    "ANTHRO 342: Theory and Practice of Cheesemongery (Graduate Seminar)",
    "CSCI 205: Advanced Programming for Cat-Lovers",
    "ENGL 112: Speculative Travel Writing"
]
```

```python
print("Course catealog reports:")
for item in courses:
    match = re.search(r"^(\w+) (\d+): (.*)$", item)
    #group(0) will return the entire string
    print("Course Dept.", match.group(1))
    print("Course #", match.group(2))
    print("Course Title", match.group(3))
```
