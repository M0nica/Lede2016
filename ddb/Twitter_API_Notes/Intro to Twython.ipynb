```python
from collections import Counter
c = Counter()
for item in all_text:
    # excludes words that are 3 letters or less - in effort to minimize stop words
    c.update(w.lower() for w in item.split() if len(w) > 3)
c.most_common(25) # shows the 25 most common words
```

```python
cursor = twitter.cursor(twitter.search, q='"steph curry", -filter:retweets', count = 100)
all_urls = list()
for tweet in cursor:
       #print(tweet['entities'])
    for item in tweet['entities']['urls']:
        all_urls.append(item['expanded_url'])
    if len(all_urls) > 100:
        break
```

```python
url_count = Counter(all_urls)
for item in url_count.most_common(10):
    print(item)
```

```python
all_media_urls = list()
for tweet in cursor:
    if 'media' in tweet['entities']:
        for item in tweet['entities']['media']:
            all_media_urls.append(item['media_url'])
    if len(all_media_urls) > 1000:
        break
```

```python
fh = open("preview.html", "w")
for item in all_media_urls:
    fh.write('<img src="{}" width="100">'.format(item))
fh.close()
```

```python
import pg8000

lakes = list()

conn = pg8000.connect(database="mondial")
cursor = conn.cursor()
cursor.execute("SELECT name, area, depth, altitude, type, river FROM lake")
for row in cursor.fetchall():
    lakes.append({'name': row[0],
                 'area': row[1],
                 'depth': row[2],
                 'elevation': row[3],
                 'type': row[4],
                 'river': row[5]})
len(lakes)
```

```python
sentences = {
    'name' : '{} is named {}'
    'area': 'The area of {} is {} square kilometers.',
    'depth': 'The depth of {} is {} meters.',
    'elevation': 'The elevation of {} is {} meters.',
    'type': 'The type of {} is "{}."',
    'river': '{} empties into a river named {}.'
}
```

```python
import random
lake = random.choice(lakes)
col = random.choice(list(lake.keys()))
print(lake)
print(col)
sentence_template = sentences[col]
```

```python
import random
def random_lake_sentence(lakes, sentences):
    rlake = random.choice(lakes)
    # get the keys in the dictionary whose value is not None; we'll only try to
    # make sentences for these
    possible_keys = [k for k, v in rlake.items() if v is not None and k != 'name']
    rframe = random.choice(possible_keys)
    return sentences[rframe].format(rlake['name'], rlake[rframe])
for i in range(10):
    print(random_lake_sentence(lakes, sentences))
```
