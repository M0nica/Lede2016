```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

# Matplotlib styles and data ink

```python
df=pd.read_csv("country-gdp-2014.csv")
```

```python
fig, ax = plt.subplots(figsize=(10,5))

#pass the ax to the .plot function 
# linewdith = 0 , removes bordr around bars
df.groupby("Continent")['life_expectancy'].median().plot(kind="barh", linewidth= 0, ax=ax, width = 0.4)
ax.set_ylabel("")
#When plotting the grid, you can sit it options
#ax.grid(True)
ax.xaxis.grid(which="major", color='MidnightBlue', linestyle=':', linewidth=1)
ax.xaxis.grid(which="minor", color='MidnightBlue', linestyle=':', linewidth=0.5)
ax.minorticks_on() # turns on minot grid -- small grid , a more frequent grid . 
#ax.yaxis.grid(color='MidnightBlue', linestyle=':', linewidth=5)
# ax.grid(color='darkgrey', linestyle='-', linewidth=1)
ax.set_axisbelow(True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

#remove ticks from the graph
plt.tick_params(
axis='x',
# which='major',
which='both',
top='off',
left = 'off',
right= 'off',
bottom='off',
labeltop='on',
labelbottom='on')

plt.tick_params(
axis='x',
# which='major',
which='minor',
top='off',
left = 'off',
right= 'off',
bottom='off',
labeltop='off',
labelbottom='off')

median = df['life_expectancy'].median()

# [-1,20 makes sure the line is long enough and fits on chart]
ax.plot([median, median], [-1,20], c='darkred', linestyle="-")
ax.annotate(s="Median life expectancy, 70 years", xy=(71,0), color="darkred")

ax.set_xlim((0,79))
```

# FiveThirtyEight scatterplot of continents' GDP and life expectancy

```python
fig, ax = plt.subplots()
df.plot(kind='scatter', x='GDP_per_capita', y='life_expectancy', ax=ax)
# no differentiation in colors between countries
```

```python
fig, ax = plt.subplots()
df[df['Continent'] == 'Africa'].plot(kind='scatter', x='GDP_per_capita', y="life_expectancy", ax=ax, color="green")
df[df['Continent'] == 'Asia'].plot(kind='scatter', x='GDP_per_capita', y="life_expectancy", ax=ax, color="red")
df[df['Continent'] == 'Oceania'].plot(kind='scatter', x='GDP_per_capita', y="life_expectancy", ax=ax)
```

```python
fig, ax = plt.subplots()
# loops through the name of every continent 
# also returns subset of dataframe

# fig.set_facecolor=('grey')
ax.set_axis_bgcolor('lightgray')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tick_params(
axis='x',
# which='major',
which='both',
top='off',
left = 'off',
right= 'off',
bottom='off',
labeltop='off',
labelbottom='on')

ax.set_prop_cycle('color',['pink','purple','red', 'deeppink', 'orange', 'darkorange'])

for continent, selection in df.groupby("Continent"):
    print(continent)
    #print(selection) # printso ut actual data
    
    # plots everything with same colors
    #selection.plot(kind='scatter', x='GDP_per_capita', y='life_expectancy', ax=ax)
    
    ax.grid(linestyle="-")
    # can set linestyle=0 or linestyle="" in order to remove lines from gaph
    ax.plot(selection['GDP_per_capita'], selection['life_expectancy'], label=continent, marker="o", linestyle="", markeredgewidth=0)

    ax.legend(loc="lower right")
    
    ax.set_ylim((30,85))
```
